```python
from matplotlib import pyplot as plt
import numpy as np
import math

x=np.arange(0, math.pi*2, 0.05)                 # 마지막 매개변수는 곡선의 매끄러움 정도이다. 데이터의 밀도? 와 관련된 듯.
y = np.sin(x)
plt.plot(x,y)                                   # 여기가 실질적으로 그래프를 생성하는 부분.
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.bar(langs,students)
plt.show()
```

```python
data = [[30,25,50,20],
        [40,23,51,17],
        [35,22,45,19]]
X = np.arange(4)
# 서로 색이 다른 각각의 그래프를 따로따로 생성.
plt.bar(X + 0.00, data[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.25)
plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=[0,25,50,75,100], color= 'g')
plt.title("histogram of result")
plt.xticks([0,25,50,75,100])                            # bins와는 달리 x축에 표시되는 값이다.
plt.xlabel('marks')
plt.ylabel('number of students')
plt.show()
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students, labels= langs, autopct='%1.2f%%')
plt.show()
```

```python
girls_grades = np.random.random(10) * 100
boys_grades = np.random.random(10) * 100
grades_range = np.arange(10,101,10)
plt.scatter(grades_range, girls_grades, color = 'r')
plt.scatter(grades_range, boys_grades, color = 'b')
plt.xlabel('Grades Range')
plt.ylabel('Grades Scored')
plt.title('Scatter plot')
plt.show()
```

```python
np.random.seed(10)                      #값은 랜덤 생성이지만 실행때마다 같은 결과가 도출된다.
dataSet1 = np.random.normal(100,30,200) # normal은 정규 분포를 따르겠다는 뜻이다. 3개의 매개변수 중 첫번째는 평균값, 두번째는 표준편차, 세번째는 추출할 샘플 수 이다.
dataSet2 = np.random.normal(80,20,200)  # 마찬가지로 평균이 80이고 표준편차가 20인 샘플 200개를 추출한다.
plotData = [dataSet1, dataSet2]
plt.boxplot(plotData)
plt.show()                              # 모르겠다 box plot은...
```
